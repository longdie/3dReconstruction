```python
## 两张图片上下拼成一张图片
import os
from PIL import Image

def combine_images(top_image_path, bottom_image_path, output_image_path):
    # 打开顶部和底部图片
    top_image = Image.open(top_image_path)
    bottom_image = Image.open(bottom_image_path)

    # 获取顶部和底部图片的宽度和高度
    top_width, top_height = top_image.size
    bottom_width, bottom_height = bottom_image.size

    # 找到两张图片中的最大宽度
    max_width = max(top_width, bottom_width)

    # 调整两张图片的宽度到最大宽度
    top_image = top_image.resize((max_width, top_height), Image.Resampling.LANCZOS)
    bottom_image = bottom_image.resize((max_width, bottom_height), Image.Resampling.LANCZOS)

    # 创建一个新的图片对象，高度为两张图片高度之和，宽度为最大宽度
    new_image = Image.new('RGB', (max_width, top_height + bottom_height))

    # 将顶部和底部图片粘贴到新图片上
    new_image.paste(top_image, (0, 0))
    new_image.paste(bottom_image, (0, top_height))

    # 保存新图片
    new_image.save(output_image_path)
    print(f"图片已合并并保存为：{output_image_path}")


root_dir = "D:\Program\ThreeD\Project\TU" # 原始数据路径
output_dir1 = "D:\Program\ThreeD\Project\easy\\train_dataset" # 保存数据路径
output_dir2 = "D:\Program\ThreeD\Project\easy\\test_dataset" 
total_num = 38

for i in range(total_num):
    id1 = i*2 + 1
    id2 = i*2 + 2
    id3 = i+1
    file1_path = os.path.join(root_dir, f"{id1}.jpg")
    file2_path = os.path.join(root_dir, f"{id2}.jpg")
    ## 因为数据量太少，测试集一次，训练集重复三次
    file3_path = os.path.join(output_dir2, f"{id3}.jpg")
    combine_images(file1_path, file2_path, file3_path)
    file3_path = os.path.join(output_dir1, f"{id3}.jpg")
    combine_images(file1_path, file2_path, file3_path)
    id3 = id3 + total_num
    file3_path = os.path.join(output_dir1, f"{id3}.jpg")
    combine_images(file1_path, file2_path, file3_path)
    id3 = id3 + total_num
    file3_path = os.path.join(output_dir1, f"{id3}.jpg")
    combine_images(file1_path, file2_path, file3_path)
```
